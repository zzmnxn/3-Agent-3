```python
from langchain_community.tools.tavily_search import TavilySearchResults
import os
from dotenv import load_dotenv
load_dotenv(override=True)
# Tavily Tool 설정
tavily_tool = TavilySearchResults(
    api_key=os.environ.get("TAVILY_API_KEY"),
    max_results=5
)

print("✅ Tavily Tool 설정 완료!")
print(f"Tool 이름: {tavily_tool.name}")
print(f"Tool 설명: {tavily_tool.description}")
```

```python
from langchain_openai import ChatOpenAI
from langchain_classic.agents import AgentExecutor, create_openai_tools_agent
from langchain_classic import hub

# 1. LLM 초기화
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# 2. Tools 리스트 생성 (Tavily Tool만 사용)
tools = [tavily_tool]

# 3. 프롬프트 가져오기 (LangChain Hub에서)
prompt = hub.pull("hwchase17/openai-functions-agent")

# 4. Agent 생성
agent = create_openai_tools_agent(llm, tools, prompt)

# 5. AgentExecutor 생성 (verbose=True로 실행 과정 확인 가능)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

print("✅ LangChain AgentExecutor 설정 완료!")
```

```python
# 테스트 1: 날씨 질문
print("=" * 60)
print("테스트 1: 내일 서울 비 와?")
print("=" * 60)
result1 = agent_executor.invoke({"input": "내일 서울 비 와?"})
print("\n답변:", result1["output"])
```

```python
# 테스트 2: 맛집 추천 질문
print("\n" + "=" * 60)
print("테스트 2: 강남역 최신 맛집 3개 추천해줘")
print("=" * 60)
result2 = agent_executor.invoke({"input": "강남역 최신 맛집 3개 추천해줘"})
print("\n답변:", result2["output"])
```

```python
# 테스트 3: 뉴스 요약 질문
print("\n" + "=" * 60)
print("테스트 3: NVIDIA 최근 뉴스 요약해줘")
print("=" * 60)
result3 = agent_executor.invoke({"input": "NVIDIA 최근 뉴스 요약해줘"})
print("\n답변:", result3["output"])
```
